```python
import matplotlib.pyplot as plt
import numpy as np
```

Analytical result

```python
h=1
def magnetization(h,beta):
    return np.tanh(beta*h)
b = np.linspace(0,10,100)
plt.plot(b, magnetization(h,b))
plt.xlabel('beta')
plt.ylabel('<m>')
plt.show()
```

Locating the transition

```python
beta = []
mag = []
for line in open('Transition_data.txt', 'r'):
    a,b,c = line.split()
    beta.append(float(a))
    mag.append(float(b))
plt.errorbar(beta, mag)
plt.xlabel('beta')
plt.ylabel('<m^2>')
plt.show()
```

Susceptibility

```python
beta = []
mag = []
error = []
for line in open('Transition_data.txt', 'r'):
    a,b,c = line.split()
    beta.append(float(a))
    mag.append(float(b))
    error.append(float(c))
plt.errorbar(beta, mag,error)
plt.xlabel('beta')
plt.ylabel('x')
plt.show()
```

Transition

```python
beta_100 = []
mag_100 = []
for line in open('Transition_data.txt', 'r'):
    a,b,c = line.split()
    beta_100.append(float(a))
    mag_100.append(float(b))
plt.plot(beta_100,mag_100)
plt.xlabel('beta')
plt.ylabel('U4 for L = 100')
plt.show()
```

```python
beta_75 = []
mag_75 = []
for line in open('Transition_data.txt', 'r'):
    a,b,c = line.split()
    beta_75.append(float(a))
    mag_75.append(float(b))
plt.plot(beta_75,mag_75)
plt.xlabel('beta')
plt.ylabel('U4 for L = 75')
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(beta_75, mag_75, label="L = 75")
ax.set_xlabel("Beta")
ax.set_ylabel("U4")
ax.plot(beta_100, mag_100, label="L = 100")
plt.legend()
plt.show()
```
